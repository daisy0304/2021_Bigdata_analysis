# 2. Line Plot
* 시간에 따른 변화의 추이를 시각화하는데 유용함
* 예) 주가 변동, 기온 변화, 아파트 가격 변동

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2-1. 그래프 그리기

```python
# (1, 2) ~ (10, 20)
x = np.arange(1, 11) # 1~10
y = x * 2 # 2~20
plt.plot(x, y, 'o')
plt.show()
```

```python
# 점과 선 둘 다 그리기
plt.plot(x, y, marker='o')
plt.show()
```

## 2-2. Marker, Line, Color
* Marker : https://matplotlib.org/stable/api/markers_api.html?highlight=marker
* Line : https://matplotlib.org/3.0.3/gallery/lines_bars_and_markers/line_styles_reference.html
* Color : https://matplotlib.org/stable/gallery/color/named_colors.html

```python
# Marker
plt.plot(x, y, marker='*', ms=10)
plt.show()
```

```python
# Line Stlye
plt.plot(x, y, ls=':')
plt.plot(y, x, ls='-.')
plt.show()
```

```python
# Color
plt.plot(x, y, color='r')
plt.plot(y, x, c='b', lw=5)
plt.show()
```

```python
# Format String
plt.plot(x, y, 'o:r')
plt.plot(y, x, '*--b')
plt.show()
```

```python
# 그래프 분리하기
plt.plot(x, y, 'o:r')
plt.figure()    # 새로운 Canvas 생성
plt.plot(y, x, '*--b')
plt.show()
```

```python
# 그래프 사이즈 변경하기 (단위 : inch)
plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.figure(figsize=(10, 5))
plt.plot(y, x)
plt.show()
```

## 2-3. Title, Label, Tick, Legend

```python
# Title 설정
plt.title('Title', fontsize=20)
plt.plot(x, y)
plt.show()
```

```python
# Label 설정
plt.xlabel('X', size=15)
plt.ylabel('Y', size=15)
plt.plot(x, y)
plt.show()
```

```python
# Tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=45)
plt.plot(x, y)
plt.show()
```

```python
# Legend(범례) 설정
plt.plot(x, y)
plt.plot(y, x)
plt.legend(['(x, y)', '(y, x)'])
plt.show()
```

```python
# Legend(범례) 설정 2
plt.plot(x, y, label='(x, y)')
plt.plot(y, x, label='(y, x)')
plt.legend()
plt.show()
```

```python
# Grid
plt.plot(x, y)
plt.grid(axis='x')
plt.show()
```

```python
# Grid 속성 변경하기
plt.plot(x, y)
plt.grid(ls=':', lw=2, c='r')
plt.show()
```

## 2-5. Subplot
* 하나의 figure에 여러 개의 plot 그리기

```python
# 1. subplot(row, column, index)
plt.subplot(1, 2, 1)
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.plot(y, x)
plt.show()
```

```python
# (1, 2, 1) -> 121
plt.subplot(211)
plt.plot(x, y)
plt.title('(x, y)')

plt.subplot(212)
plt.plot(y, x)
plt.title('(y, x)')

plt.suptitle('Title', size=20)
plt.tight_layout()  # 자동 레이아웃 설정
plt.show()
```

```python
# 2. add_subplot()
fig = plt.figure()
axes1 = fig.add_subplot(2, 1, 1)
axes2 = fig.add_subplot(2, 1, 2)

axes1.plot(x, y)
axes1.set_title('(x, y)')

axes2.plot(y, x)
axes2.set_title('(y, x)')
plt.show()
```

```python
# 3. subplots()
fig, axes = plt.subplots(2, 1)  # row, col

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
plt.show()
```

## 2-7. 한글 출력하기

```python
# 설치된 글꼴 찾기
from matplotlib import font_manager

for font in font_manager.fontManager.ttflist:
    if 'Nanum' in font.name:
        print(font.name, font.fname)
```

```python
# 1. 전역으로 설정하기
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['font.size'] = 20
plt.plot(x, y)
plt.title('타이틀')
plt.show()
```
