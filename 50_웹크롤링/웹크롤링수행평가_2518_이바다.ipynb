# 유튜브 웹크롤링
* 유튜브 웹크롤링
* 영상제목, 조회수, 업로드 날짜, 영상 길이 추출

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver_path = 'C:\\JupyterLab\\chromedriver.exe'
driver = webdriver.Chrome(driver_path)

driver.get("https://www.youtube.com/c/BANGTANTV/videos")
```

```python
soup = BeautifulSoup(driver.page_source, 'lxml')
```

## 영상제목 추출

### 추출방식
* soup.select('a#video-title')를 통해 video 정보를 가져온다.
* title 리스트를 만들어, video 제목만 가져와 리스트에 담는다.

```python
videos = soup.select('a#video-title')
title = []

for video in videos:
    title.append(video.text.strip())
title
```

## 조회수 추출

### 추출방식
* soup.select('span.style-scope.ytd-video-meta-block') 명령어로 'span' 태그를 가지고 있고, 클래스는 style-scope.ytd-video-meta-block 가진 정보를 추출한다.
* 업로드 날짜와 조회수를 구분하기 위해 자식태그를 '1'로 설정하여 조회수만 뽑는다.

```python
soup.select('span.style-scope.ytd-video-meta-block')
view = soup.select('div#metadata-line > span:nth-child(1)')
view
```

## 업로드 날짜 추출

### 추출방식
* soup.select('span.style-scope.ytd-video-meta-block') 명령어로 'span' 태그를 가지고 있고, 클래스는 style-scope.ytd-video-meta-block 가진 정보를 추출한다.
* 업로드 날짜와 조회수를 구분하기 위해 자식태그를 '2'로 설정하여 업로드 날짜만 뽑는다.

```python
soup.select('span.style-scope.ytd-video-meta-block')
date = soup.select('div#metadata-line > span:nth-child(2)')
date
```

## 영상길이 추출

### 추출방식
* 'span' 태그를 가지고 클래스로 style-scope ytd-thumbnail-overlay-time-status-renderer를 가지는 정보를 추출한다.
* 영상 길이를 추출하기 위해 strip 명령어를 사용하고, 중복되는 시간 표현을 없애기 위해 '\n'을 기준으로 잘라 앞 부분만 사용하여 리스트에 넣는다.

```python
lengths = soup.find_all('span', 'style-scope ytd-thumbnail-overlay-time-status-renderer')
length = []

for time in lengths:
    length.append(time.text.strip().split('\n')[0])
length
```

### 각 데이터를 리스트에 넣어 정리한다.

```python
title_list = []
date_list = []
view_list = []
length_list = []

for i in range(len(title)):
    title_list.append(title[i])
    date_list.append(date[i].text.strip())
    view_list.append(view[i].text.strip())
    length_list.append(length[i])
    
print(title_list)
print(date_list)
print(view_list)
print(length_list)
```

## 영상 제목, 조회수, 업로드 날짜, 영상 길이로 DataFrame 만들기

```python
df_data = {
    '영상 제목' : title,
    '조회수' : view,
    '업로드 날짜' : date,
    '영상 길이' : length
    
}

df = pd.DataFrame(data=df_data)
df
```
