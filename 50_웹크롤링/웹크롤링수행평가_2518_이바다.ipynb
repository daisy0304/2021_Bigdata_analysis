```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver_path = 'C:\\JupyterLab\\chromedriver.exe'
driver = webdriver.Chrome(driver_path)

driver.get("https://www.youtube.com/c/BANGTANTV/videos")
```

* 'span' íƒœê·¸ë¥¼ ê°€ì§€ê³  í´ë˜ìŠ¤ë¡œ style-scope ytd-thumbnail-overlay-time-status-rendererë¥¼ ê°€ì§€ëŠ” ì •ë³´ë¥¼ ì¶”ì¶œí•œë‹¤.

```python
lengths = soup.find_all('span', 'style-scope ytd-thumbnail-overlay-time-status-renderer')
length = []

for time in lengths:
    length.append(time.text.strip().split('\n')[0])
length
```

### ê° ë°ì´í„°ë¥¼ ë¦¬ìŠ¤íŠ¸ì— ë„£ì–´ ì •ë¦¬í•œë‹¤.

```python
title_list = []
date_list = []
view_list = []
length_list = []

for i in range(len(title)):
    title_list.append(title[i])
    date_list.append(date[i].text.strip())
    view_list.append(view[i].text.strip())
    length_list.append(length[i])
    
print(title_list)
print(date_list)
print(view_list)
print(length_list)
```
