```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ver3.csv')
df.head()
```

```python
df.columns = ['time','gender','age','annoy','frequency','where','realize','express_example','why','haveto']
df.info()
```

```python
df = df.fillna('-')
```

```python
def gender_change(x):
    if x == '여성' : return 1
    else : return 0

df['gender'] = df['gender'].apply(gender_change)
df.info()
```

```python
def age_change(x):
    return int(x[:-1])

df['age'] = df['age'].apply(age_change)
df.head()
```

```python
#    def annoy_change(x):
#        x = x.split(',')
#        for item in x :
#            item = item[0:6]
#        return x
#        
#    df['annoy'] = df['annoy'].apply(annoy_change)
```

```python
def OX_change(x):
    if x == '네' : return 1
    else : return 0

df['frequency'] = df['frequency'].apply(OX_change)
df['realize'] = df['realize'].apply(OX_change)
df.info()
```
